## Criar um gráfico de histograma para visualizar a distribuição da massa dos pinguins

```python
# Importando as bibliotecas
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Carregando o Conjunto de dados
df = sns.load_dataset("penguins")
df.head()
```

```python
# Criando um gráfico do tipo HISTOGRAMA
ax = sns.histplot(data=df, x='body_mass_g')
```

```python
ax = sns.histplot(data=df, x='body_mass_g')

# Utilizando linhas Verticais e Horizontais
sns.set_style("whitegrid")
```

```python
ax = sns.histplot(data=df, x='body_mass_g')

sns.set_style("whitegrid")

# Melhorando o título e os rótulos dos eixos
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')
```

```python
ax = sns.histplot(data=df, x='body_mass_g')

sns.set_style("whitegrid")

plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')

# Adicionando Linha à origem e Removendo as bordas do grid
ax.set(frame_on=False)
ax.axhline(0, color='k', clip_on=False)
plt.show()
```
