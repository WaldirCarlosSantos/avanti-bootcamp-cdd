## Criar um gráfico de barras simples para visualizar a arrecadação total para o conjunto de arrecadação de filmes

```python
# Importando as bibliotecas
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Lendo e carregando o arquivo no formato csv
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

```python
# Plotando o gráfico de barras na horizontal
ax = sns.barplot(data = df, y = 'titulo', x = 'arrecadacao')
```

```python
ax = sns.barplot(data = df, y = 'titulo', x = 'arrecadacao')

# Utilizando Valores Simples - Divisão dos valores da arrecadaçaõ por 1 milhão
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)
```

```python
# Ordenando as Barras na ordem decrescente
df = df.sort_values(by = 'arrecadacao', ascending = False)
```

```python
# Plotando o gráfico de barras na horizontal após ordenar as barras
ax = sns.barplot(data = df, y = 'titulo', x = 'arrecadacao')
```

```python
# Plotando o gráfico de barras na horizontal após adicionar linhas de grade
ax = sns.barplot(data = df, y = 'titulo', x = 'arrecadacao')

# Utilzando linhas de grade para facilitar a leitura comparativa de proporções entre as barras
sns.set_style('whitegrid')
```

```python
# Plotando o gráfico de barras na horizontal, ajustando o tamanho e mudando a cor das barras
plt.figure(figsize=(5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

sns.set_style('whitegrid')

# Adicionando valores nas barras
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type ='center',
    color='white'
)
```

```python
plt.figure(figsize=(5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

sns.set_style('whitegrid')

ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type ='center',
    color='white'
)

# Melhorar o título e os labels dos eixos
plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')
```

```python
plt.figure(figsize=(5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

sns.set_style('whitegrid')

ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type ='center',
    color='white'
)

plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')

# Adicionar uma linha vertical na origem e remover bordas
ax.set(frame_on=False)
ax.axvline(0, color = 'k', clip_on = False)
plt.show()
```
