## Criar um gráfico de caixas para visualizar a quantidade total da conta em relação ao dia da semana

```python
# Importação das bibliotecas
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Carregando o Conjunto de dados
df = sns.load_dataset("tips")
df.head()
```

```python
# Criando um gráfico do tipo BOXPLOT
ax = sns.boxplot(x='day', y='total_bill', data=df)
```

```python
ax = sns.boxplot(x='day', y='total_bill', data=df)

# Adicionando linhas horizontais para facilitar a comparação de proporções entre os boxes
sns.set_style("whitegrid")
```

```python
# Ajustando o tamanho da figura e utilizando a mesma cor para as barras
plt.figure(figsize=(6,3))
ax = sns.boxplot(x='day', y='total_bill', data=df, color='cornflowerblue')

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(6,3))
ax = sns.boxplot(x='day', y='total_bill', data=df, color='cornflowerblue')

sns.set_style("whitegrid")

# Melhorando o título e os rótulos dos eixos
plt.xlabel('Dia da Semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
