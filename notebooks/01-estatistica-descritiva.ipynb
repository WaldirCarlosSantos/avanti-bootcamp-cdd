```python
quantitative_vars = ['total_bill', 'tip', 'size']
```

### 3.3.4 - Media, Mediana, Desvio Padrão, Mínimo, Máximo das Variáveis QUANTITATIVAS

```python
media = df[quantitative_vars].mean()
mediana = df[quantitative_vars].median()
desvio_padrao = df[quantitative_vars].std()
minimo = df[quantitative_vars].min()
maximo = df[quantitative_vars].max()

print('MÉDIA')
print(media)
print()

print('MEDIANA')
print(mediana)
print()

print('DESVIO PADRÃO')
print(desvio_padrao)
print()

print('MÍNIMO')
print(minimo)
print()

print('MÁXIMO')
print(maximo)
```

## 3.4. Calculando Estatísticas para cada Variável Categórica (QUALITATIVA)

```python
# Chamando as variáveis do tipo Qualitativa
qualitative_vars
```

```python
# Apresentando os Valores presentes em cada uma das Variáveis
genero_sexual = df.sex.unique()
print('Gênero Sexual')
print(genero_sexual)
print()

fumantes = df.smoker.unique()
print('Fumantes')
print(fumantes)
print()

dias_da_semana = df.day.unique()
print('Dias da Semana')
print(dias_da_semana)
print()

refeicao = df.time.unique()
print('Refeição')
print(refeicao)
print()
```

### 3.4.1 - Estatísticas Gerais para cada variável QUALITATIVA


```python
# Exemplo:
# Da variável 'sex', apenas para o valor "Female", apresentar todos os dados
df.query('sex == "Female"').describe()
```

```python
# Criando uma tabela completa de Estatísticas Gerais para cada Valor Único presente em cada Variável Qualitativa
for variavel in qualitative_vars:
    valores_unicos = df[variavel].unique()
    for valor in valores_unicos:
        print(f"{variavel} - {valor}")
        subamostra = df[df[variavel] == valor]
        # print(subamostra.describe())
        display(subamostra.describe())
        print("=" * 100)
```
